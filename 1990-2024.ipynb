```python
import pandas as pd

# Assuming your DataFrame is named 'df'
df = pd.read_csv("C:/Users/mania/OneDrive/Desktop/Econophysics/Data/Nifty100(1990-2024).csv")

# Calculate the price change using the 'diff' function
df['Price Change'] = df['Close'].diff()
df['Log_Close'] = np.log(df['Close'])

# Handle the first row (which will have NaN due to no previous value)
df['Price Change'].fillna(0, inplace=True)

# Calculate the percentage change
df['Percentage Change'] = df['Close'].pct_change() * 100

# Handle the first row's NaN value for percentage change
df['Percentage Change'].fillna(0, inplace=True)
df['Log_Change'] = np.log(df['Percentage Change'])


# Display the DataFrame with the new columns
print(df.head())
```

```python
# Plot the percentage change against the date
df.plot(x='Date', y='Percentage Change')

# Add labels and titles
plt.xlabel('Date')
plt.ylabel('Percentage Change')
plt.title('Percentage Change of Nifty100')

# Show the plot
plt.show()
```

```python
# Plot the percentage change against the date
df.plot(x='Date', y='Close')

# Add labels and titles
plt.xlabel('Date')
plt.ylabel('Close')
plt.title('Close of Nifty100')

# Show the plot
plt.show()
```

```python
# Plot the percentage change against the date
df.plot(x='Date', y='Log_Close')

# Add labels and titles
plt.xlabel('Date')
plt.ylabel('Log Close')
plt.title('Log change of Nifty')

# Show the plot
plt.show()
```
